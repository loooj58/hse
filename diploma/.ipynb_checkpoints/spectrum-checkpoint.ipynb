```julia
calib_coeff = []
sigma_coeff = []
for i in range(0, stop=19)
    f = open("calibration/sigma_coefficient_puck_$(i).txt")
    lines = map(x->parse(Float64, x), split(readlines(f)[1]))
    close(f)
    println(i, lines)
    append!(calib_coeff, lines[1])
    append!(sigma_coeff, lines[2])
end

println(calib_coeff)
println(sigma_coeff)
```

```julia
function get_sigma(puck_number, value)
    return sqrt(value * calib_coeff[puck_number]) * sigma_coeff[puck_number]
end

function get_energy(puck_number, value)
    return value * calib_coeff[puck_number]
end
```

```julia
data = Dict()
for i in range(0, stop=9)
    f = open("electron_processed_data/electron_data_from_puck_$(i).txt")
    lines = map(x->get_energy(i+1, parse(Float64, x)), readlines(f))
    close(f)
#     println(i, lines)
    data[i] = lines
end
```

```julia
data
```

```julia
data_integral = [sum(value) for (key, value) in data]
```

```julia
using PyPlot
plot(data_integral)
```

```julia
data_differ = zeros(length(data[1]))
for (key, value) in data
    for i in range(1, stop=length(value))
        data_differ[i] += value[i]
    end
end
```

```julia
hist(data_differ, bins=100)
print()
```
