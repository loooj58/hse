```julia
sig = alpha * Omega
sig_inv = safe_inv(sig)
data_errors_inv = safe_inv(data_errors);
```

```julia
using Distributions

likelihood = let K=K, data_errors_inv=data_errors_inv, f=f
    phi -> begin
        mu = K * phi
        return -1/2 * transpose(f - mu) * data_errors_inv * (f - mu)
    end
end
    
prior = let sig_inv=sig_inv
    phi -> begin
        mu = zeros(length(phi))
        return -1/2 * transpose(phi - mu) * sig * (phi - mu)
    end
end
```
